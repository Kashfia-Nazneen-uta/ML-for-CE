#### Compute a new matrix $J$ from $H$, where 1) The negative numbers are replaced by their absolute value, and 2) The positive elements are replaced by 10.
- Use a ```for``` loop to iterate over each (i, j) index pair.
- Use np.where

````python
import numpy as np

np.random.seed(1) 
J = np.random.normal(loc=0, scale=1, size=(5,5)) # loc and scale can be ignored if it is standard normal distribution
H = J.copy()

# 1. Use for loop
len_H = len(H)
for i in range(len_H):
    for j in range(len_H):
        if H[i][j] < 0:
            J[i][j] = abs(H[i][j])
        elif H[i][j] > 0:
            J[i][j] = 10
        else:
            pass

print('J:\n', J)

# or
len_H = len(H)
for i in range(len_H):
    for j in range(len_H):
        if H[i][j] <= 0:
            J[i][j] = abs(H[i][j])
        else:
            J[i][j] = 10

print('J:\n', J)
````

````python
# 2. Use np.where loop
# hint for calculating absolute values using np.abs(): np.abs(-3) = 3
J = np.where(H<=0, np.abs(H), 10)
print('J:\n', J)
````

## Comprehension

#### Produce a new list name 'H_list' by flattening $H$ into a 1d array and convert it to a list
Hint: Use [numpy.ndarray.flatten](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.flatten.html) as well as
[numpy.ndarray.tolist](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.tolist.html).

#### Use list comprehension to modify 'H_list' as follows: 1) replace negative number by their absolute value, and 2) replace positive numbers by 10. Return 'H_list_new' after all operations.

## Functions

#### Write a function ```sum_square(n)``` that takes as input $n$ and returns $y$, where $$y=\left(1+2^{2}+\cdots+n^{2}\right)=\sum_{k=1}^{n}k^{2}$$
- Use for loop
- Use np.sum()

````python
# using a for loop
# write function 'sum_square(n)'
def sum_square(n):
    y = 0
    for i in range(1, n+1):
        y += i**2
    return y

# apply function
print('Sum of squares from 1 to 5:', sum_square(5))
````

````python
# using np.sum()
# a regular function
def sum_square(n):
    return np.sum(np.arange(1,n+1)**2) 

# a lambda function
y = lambda n: np.sum(np.arange(1,n+1)**2) 

print('Sum of squares from 1 to 5 using np.sum() with regular function:', sum_square(5))
print('Sum of squares from 1 to 5 using np.sum() with lambda function:', y(5))
````

- Plot $sin(x)$ and $sec(x)$ between $-3\pi$ and $3\pi$ on the same graph, using different colors, line styles, and markers for the two curves. 
- Include a legend to distinguish between the two curves. 
- Add $x$ and $y$ labels, and a title to the plot. 
- You will probably need to use the function ```ylim```, which set the range of $y$ values, to enhance the readability of your plot.
- save the plot as a pdf (dpi=250; tight layout) to the a neighboring folder named 'plot_folder'. Relative path should be used. Hint: the path looks something like './plot_folder/{plot_name}.pdf'

For those who are interested, here is an article about [making publication quality plot using Python](https://github.com/jbmouret/matplotlib_for_papers).

````python
# import necessary packages
import numpy as np
import matplotlib.pyplot as plt
import os
````

````python
# make and format plot
# define x-values
# this creates a vector of numbers on the interval [-3*pi, 3*pi), incremented by 0.1
x_values = np.arange(-3*np.pi, 3*np.pi, 0.1)

# calculate y-values as the sine of each x-value
sin_values = np.sin(x_values)         # np.sin(...) performs calculation element-wise
sec_values = 1/np.cos(x_values)      # secant values, which is 1/cosine

# plot results
# define the figure size with resolution 200 pixels per inch
plt.figure(figsize=(8, 6), dpi=100)
# plot sin values
plt.plot(x_values, sin_values,'b-',marker = 'o')    # style is 'b-', meaning blue solid line with marker 'o'
# plot sec values
plt.plot(x_values, sec_values,'r--',marker = 'x')    # style is 'r-', meaning red dashed line with marker 'x'
# add a title to the plot
plt.title("sin(x) and sec(x) function from -3*pi to 3*pi")
plt.xlabel('x [rad]')
plt.ylabel('Function value')
plt.grid()
plt.legend(['sin(x)', 'sec(x)'])
# use ylim to set y-axis limits to enhance readability
plt.ylim(-5, 5)
# use tight_layout to make sure plot elements fit within the figure area
plt.tight_layout()
````

````python
# save the plot
out_dir = "./plot_folder"
os.makedirs(out_dir, exist_ok=True)
plt.savefig('./plot_folder/sine_and_sec_function.pdf', dpi=250)
plt.show()
````

#### Describe a [*magic number*](https://goo.gl/nYMuuG) and the benefits of avoiding 'Unnamed numerical constants' in your code.

**Magic number**

A magic number is a direct usage of a number in the code without any explanation or name.

**Benefits of avoiding unnamed numerical constants**

If an 'unnamed numerical constant' or magic number is avoided by replacing it with a named constant, there are several benefits, such as:
- It improves clarity as the name suggests the meaning of the constant
- Same constant can be reused multiple times throughout the code without introducing errors.
- It makes the code easier to update. Changing the value of a named constant requires updating it in only one location rather than searching for and modifying multiple occurrences. For example, a constant MAX_SPEED can be used several times in a code. If we used a named constant, we only need to change the value in one place, rather than searching and updating the value in many places. 
- It reduces the risk of inconsistent updates, which can lead to bugs within the code.
